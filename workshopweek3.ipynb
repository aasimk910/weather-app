```python
import matplotlib.pyplot as plt

days = ["DAY 1", "DAY 2", "DAY 3", "DAY 4", "DAY 5"]
enfield_distance = [80, 40, 70, 80, 20]
honda_distance = [50, 20, 20, 50, 60]
yamaha_distance = [70, 20, 60, 40, 60]
ktm_distance = [40, 20, 30, 70, 60]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(days, enfield_distance, marker='o', label='Enfield')
plt.plot(days, honda_distance, marker='o', label='Honda')
plt.plot(days, yamaha_distance, marker='o', label='Yamaha')
plt.plot(days, ktm_distance, marker='o', label='KTM')

# Set labels for the x and y axes
plt.xlabel("Days")
plt.ylabel("Distance Covered (kms)")

# Set title for the plot
plt.title("Bike Details in Plot")

# Display the plot
plt.grid(True)
plt.show()
```

```python
#Example 5:
import matplotlib.pyplot as plt
#create a function to calculate y values
def function(x):
  result= 3*x+2
  return result
#defining empty lists
xCord=[]
yCord=[]
#loop over 10 values from 0 to 9, and calculate the matching values for y using the function
for i in range(0,10):
    xCord.append(i)
    yCord.append(function(i))
plt.plot(xCord,yCord, marker='x',color='red')
#function to add title
plt.title('Example 5')
#Function to label axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# function for the visibility of grid
plt.grid()
plt.show()
```

```python
#exercise3
import matplotlib.pyplot as plt
import math

x1 = [-2 + i * 4 / 19 for i in range(20)]
x2 = [-3 + i * 6 / 19 for i in range(20)]
x3 = [-10 + i * 20 / 19 for i in range(20)]


def function1(x):
    return 5 * x**3 + 2 * x - 1


def function2(x):
    return -2 * x**3 + x**2 + 100


def function3(x):
    return 2 * math.pi * x + 20


y1 = [function1(x) for x in x1]
y2 = [function2(x) for x in x2]
y3 = [function3(x) for x in x3[::]]

plt.plot(x1, y1, color='blue', label='f(x) = 5x^3 + 2x - 1')
plt.plot(x2, y2, color='red', label='f(x) = -2x^3 + x^2 + 100')
plt.plot(x3, y3, color='green', label='f(x) = 2x + 20')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Three Functions', fontsize=16, fontweight='bold')

plt.legend()
plt.grid()
plt.show()
```

```python
#exercise4

import matplotlib.pyplot as plt
import math

x1 = [2 + i * 4 / 19 for i in range(20)]
x2 = [3 + i * 6 / 19 for i in range(20)]
x3 = [10 + i * 20 / 19 for i in range(20)]


def function1(x):
    return 5 * x**3 + 2 * x - 1


def function2(x):
    return -2 * x**3 + x**2 + 100


def function3(x):
    return 2 * (22 / 7) * x + 20


y1 = [function1(x) for x in x1]
y2 = [function2(x) for x in x2]
y3 = [function3(x) for x in x3]

plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.plot(x1, y1, label='f(x) = 5x^3 + 2x - 1', color='blue')
plt.title('Function 1')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x2, y2, label='f(x) = -2x^3 + x^2 + 100', color='red')
plt.title('Function 2')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x3, y3, label='f(x) = 2x + 20', color='green')
plt.title('Function 3')
plt.legend()

plt.suptitle('Plots of Three Functions', fontsize=16, fontweight='bold')

plt.show()
```

```python
#exercise5
import matplotlib.pyplot as plt

days = ['DAY 1', 'DAY 2', 'DAY 3', 'DAY 4', 'DAY 5']
enfield = [50, 40, 70, 80, 20]
honda = [80, 20, 20, 50, 60]
yahama = [70, 20, 60, 40, 60]
ktm = [40, 20, 30, 70, 60]

plt.subplot(2, 2, 1)
plt.plot(days, enfield, marker='o', color='red')
plt.title('ENFIELD')
plt.xlabel('Days')
plt.ylabel('Distance Covered (km)')

plt.subplot(2, 2, 2)
plt.plot(days, honda, marker='x', color='green')
plt.title('HONDA')
plt.xlabel('Days')
plt.ylabel('Distance Covered (km)')

plt.subplot(2, 2, 3)
plt.plot(days, yahama, marker='^', color='blue')
plt.title('YAHAMA')
plt.xlabel('Days')
plt.ylabel('Distance Covered (km)')

plt.subplot(2, 2, 4)
plt.plot(days, ktm, marker='+', color='yellow')
plt.title('KTM')
plt.xlabel('Days')
plt.ylabel('Distance Covered (km)')

plt.tight_layout()

plt.show()
```
